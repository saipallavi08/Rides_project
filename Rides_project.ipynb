```python
import plotly.express as px
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1=pd.read_csv("/content/rides.csv")
```

```python
#printing original data
data1
```

```python
#above data havinng null values
#now removeing null values
```

```python
df=data1.dropna()
data=data1.dropna()
newdata=data1.dropna()
```

```python
df.head()
```

```python
plt.figure(figsize=(10, 8))
plt.style.use('fivethirtyeight')
plt.title("Distribution of  Rides")
sns.distplot(data['Drivers Active Per Hour'])
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.style.use('fivethirtyeight')
plt.title("Distribution of Riders ")
sns.distplot(data['Riders Active Per Hour'])
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.style.use('fivethirtyeight')
plt.title("Distribution of Rides Completed ")
sns.distplot(data['Rides Completed'])
plt.show()
```

```python
a = data["Drivers Active Per Hour"].sum()
b = data["Riders Active Per Hour"].sum()
c = data["Rides Completed"].sum()
labels = ['Drivers per','Rider per hour','Rides completed']
values = [a,b,c]

fig = px.pie(data, values=values, names=labels,
             title='rider per hour driver per hour rides completed', hole=0.5)
fig.show()
```

```python
violin_plot= px.violin(data, x='Riders Active Per Hour', y='Drivers Active Per Hour',
                        title='Distribution of Supply Ratio across Different Levels of Riders Active Per Hour')
violin_plot.update_xaxes(title='Riders Active Per Hour')
violin_plot.update_yaxes(title='Supply Ratio')
violin_plot.show()
```

```python
scatter_plot = px.scatter(data, x='Riders Active Per Hour', y='Rides Completed', title='Relationship between Riders Active Per Hour and Rides Completed')
scatter_plot.update_xaxes(title='Riders Active Per Hour')
scatter_plot.update_yaxes(title='Rides Completed')
scatter_plot.show()
```
